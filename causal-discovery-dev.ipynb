```python
import sys
sys.path.append('backend/fastlingam')

import networkx as nx
import numpy as np
import os
import sys
import lingam
import matplotlib.pyplot as plt

output_directory = "./causal_graphs"
os.makedirs(output_directory, exist_ok=True)

data_filename = "./data/sp100.csv"
raw_data = np.genfromtxt(data_filename, delimiter=',', skip_header=1)
data = (raw_data[1:] / raw_data[:-1]) - 1
# data = raw_data
with open(data_filename, 'r') as f:
    header = f.readline().strip().split(',')
stock_symbols = header
```

```python
def calculate_summary_matrix(data):
    model = lingam.VARLiNGAM(lags=2, prune=False)
    model.fit(data)
    # summary_matrix = np.sum(np.abs(model.adjacency_matrices_[1:]), axis=0)
    summary_matrix = np.abs(model.adjacency_matrices_[0])
    # summary_matrix = np.sum(np.abs(model.adjacency_matrices_), axis=0)
    np.fill_diagonal(summary_matrix, 0)
    return summary_matrix
```

```python
window_size = 500
nsamples = data.shape[0]
step_size = 1
for i in range(window_size, nsamples, step_size):
    fn = './matrices/frame-' + str(i).zfill(4) + '.csv'
    print(fn)
    sm = calculate_summary_matrix(data[i-window_size:i,:])
    np.savetxt(fn, sm, delimiter=",")
```
